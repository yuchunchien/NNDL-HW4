#### Bullet 5: Observation on embedding pairs

##### Preparation:

```python
from hw4a import test_rnnslu
rnn, dic = test_rnnslu()
```

```python
idx2word = dict((k, v) for v, k in dic['words2idx'].items())
print(idx2word)
```

```python
A = rnn.get_embed()
for i in range(A.shape[0]):
    print("%i: ") % (i) 
    print(A[i])
```

```python
import numpy
from numpy import random

B=[]
C=[]

for t in range(10000):
    i = int(random.uniform(0,A.shape[0]-1))
    j = int(random.uniform(i+1,A.shape[0]-1))
    k = int(random.uniform(j+1,A.shape[0]-1))
    l = int(random.uniform(k+1,A.shape[0]-1))
        
    if(i==A.shape[0]-1 or j==A.shape[0]-1 or k==A.shape[0]-1 or l==A.shape[0]-1):
        continue
        
    B.append([i,j,k,l,numpy.sum(((A[i]-A[j]) - (A[k]-A[l]))**2)])
    C.append(numpy.sum(((A[i]-A[j]) - (A[k]-A[l]))**2))
```

```python
for t in range(len(B)):
    print("%i:") % t
    print(B[t])
```

```python
min_c = min(C)
min_c_idx = C.index(min_c)

print(min_c)
print(min_c_idx)
```

##### Discuss:

The list contains [i,j,k,l,diff], while i,j,k,l are index of the chosen word.

The list represents (F(i) - F(j)) - (F(k) - F(l)) = diff.

According to my inspection, when diff is lower than 1.4, the list can be categorized as "having special relations between word pairs."

1.
[198, 568, 569, 571, 1.1491215] -> (F('fn') - F('yn')) - (F('york') - F('your')) = 1.1491215

2.
[546, 547, 549, 550, 1.1103785] -> (F('wednesday') - F('wednesdays')) - (F('weekday') - F('weekdays')) = 1.1103785

3.
[546, 416, 547, 417, 1.2820058] -> (F('wednesday') - F('saturday')) - (F('wednesdays') - F('saturdays')) = 1.2820058

4.
[496, 547, 566, 568, 1.0760145] -> (F('thursday') - F('wednesdays')) - (F('y') - F('yn')) = 1.0760145

5.
[566, 568, 569, 571, 1.3039891] -> (F('y') - F('yn')) - (F('york') - F('your')) = 1.3039891
